### Task 1
In a single plot, plot the following:-
1. Plot Sin wave from 0-360 degrees. (Degrees on X axis and value of sin on Y axis)
2. Plot Cos wave from 45-360 degrees. (Degrees on X axis and value of cos on Y axis)
3. Add legend to the plot
4. Add Axis Labels
5. Add Plot Title as "Trigonometric Plots"
6. Add gridlines to the plot

```python
#importing the necessary libraries
import matplotlib.pyplot as plt  
import numpy as np  
#1.
#setting the values of x-axis and y-axis using numpy.
x = np.arange(0,2*np.pi+0.1,0.1)             #setting x plane startting point for sine from 0 degree.
xd = np.degrees(x)# start,stop,step          # converting the radian value into degree.
y = np.sin(x)
r = np.arange(0.25*np.pi,2*np.pi+0.1,0.1)    #setting x plane startting point for cose from 45 degrees.
rd = np.degrees(r)                            # converting the radian value into degree.

# 2.
z = np.cos(r)
plt.plot(xd,y,rd,z)                        # plotting the graph

# 3.
plt.legend(["sin", "cos"], loc = "lower left")         #setting legend

# 4.
plt.ylabel("Values of sin and cos")
plt.xlabel("Degrees")                   

# 5.
plt.title("Trigonometric Plots")

# 6.
plt.grid()
plt.show()
```

### Task 2
The dataset "candy-data-clean.csv" contains data for different candy / chocolate bars.
The data contains a feature named "click", this feature records that the candy / chocolate bar advertisement was clicked or not.

Perform the following:-
1. Import the dataset
2. Use the "click" feature as dependent variable / output
3. Perform Exploratory Data Analysis and make atleast 3 inference about the dataset using visual plots
4. Using logistic regression, use other features to find out the values for the "click" variable
5. Create a dataframe comparing actual values of "click" and predicted values of "click"
6. Calculate the performance metrics for the model


```python
# 1.
import pandas as pd
import seaborn as sns
df = pd.read_csv("candy-data-clean.csv")
df.head()
df.drop(['competitorname', 'sugarpercent', 'pricepercent'], axis = 1, inplace = True)

# 2.
X = df.iloc[:, :-1].values
y = df.iloc[:,-1].values
```

```python
# 3.
# chocolate or not chocolate
sns.set_style('darkgrid')
sns.countplot(x = 'click',hue = 'chocolate', data = df)
```

```python
sns.set_style('whitegrid')
sns.countplot(x = 'click', hue = 'fruity',data = df)
```

```python
sns.set_style('whitegrid')
sns.countplot(x = 'click', hue = 'caramel',data = df)
```

```python
# 4.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20, random_state = 10)
from sklearn.linear_model import LogisticRegression
regression = LogisticRegression()
regression.fit(X_train, y_train)
predictions = regression.predict(X_test)
```

```python
# 5.
compare = pd.DataFrame({"Actual":y_test, "Predicted":predictions})
compare
```

```python
# 6.
from sklearn.metrics import classification_report, confusion_matrix
cnf_matrix = confusion_matrix(y_test, predictions)
print(classification_report(y_test, predictions))
```
